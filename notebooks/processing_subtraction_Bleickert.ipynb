```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('../datasets/images/test.jpg', 0)
```

```python
#subtraction

none = cv2.imread('../datasets/images/none.jpg',0)
imgs = (img-30-none)
_, imgst = cv2.threshold(imgs, 50, 255, cv2.THRESH_BINARY)
plt.imshow(imgst-255, cmap = 'gray')
plt.title('subtraction')
```

```python
#subtraction with equalisation

imge = cv2.equalizeHist(img)
nonee = cv2.equalizeHist(none)
imgse = (imge-30-nonee)
_, imgset = cv2.threshold(imgse, 50, 255, cv2.THRESH_BINARY)
plt.imshow(imgset-255, cmap = 'gray')
plt.title('subtraction with equalisation')
```

```python
#subtraction and dilation

kernel = np.ones((2, 1), np.uint8)
imgsd = cv2.dilate(imgset, kernel, iterations = 8)
plt.imshow(imgsd-255, cmap = 'gray')
plt.title('subtraction and dilation')

imgd2 = cv2.bitwise_not(imgsd)
cv2.imwrite('result.jpg', imgd2)
```

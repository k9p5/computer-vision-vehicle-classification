```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('../datasets/images/test.jpg', cv2.COLOR_BGR2GRAY)
```

```python
#canny edge detector

imgc = cv2.Canny(img, 60, 250)
plt.imshow(imgc, cmap = 'gray')
```

```python
#canny edge detector trackbars

def nothing(x):
    pass
cv2.namedWindow("canny")
cv2.createTrackbar("t_high", "canny", 250, 2000, nothing)
cv2.createTrackbar("t_low", "canny", 60, 2000, nothing)
imgt=img.copy()
k=0
while(k!=27):
    cv2.imshow('canny',imgt)
    k = cv2.waitKey(1) & 0xFF
    th = cv2.getTrackbarPos('t_high','canny')
    tl = cv2.getTrackbarPos('t_low','canny')
    imgt=cv2.Canny(img, tl, th)
```

```python
#canny edge detector with subtraction

imgc2 = cv2.Canny(img, 200, 200)
none = cv2.imread('../datasets/images/none.jpg', cv2.COLOR_BGR2GRAY)
nonec = cv2.Canny(none, 10, 10)

imgcs = (imgc2-nonec)

plt.imshow(imgcs, cmap = 'gray')
```

```python
#canny edge detector with subtraction and dilation

kernel = np.ones((2,1),np.uint8)
imgcsd = cv2.dilate(imgcs,kernel,iterations = 1)
plt.imshow(imgcsd, cmap = 'gray')
```

```python
#canny edge detector with subtraction and morphology

kernel = np.ones((1,2),np.uint8)
imgcsm = cv2.morphologyEx(imgcs, cv2.MORPH_OPEN, kernel)
plt.imshow(imgcsm, cmap = 'gray')

#imgd2 = cv2.bitwise_not(imgcs)
cv2.imwrite('result.jpg', imgcsm)
```
